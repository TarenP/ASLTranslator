```python
ep = 8.854*(10**-12) #coloumb's constant
k_constant = 0.5 # some constant for charge density 
def shell_E_field(r, a, b):
    """returns E-field at the given position given the thickness of the shell"""
    
    if (r < a):
        return 0
    elif (r >= a and r < b):
        return (k_constant*(r-a))/(ep*(r**2))
    elif (r >= b): 
        return (k_constant*(b-a))/(ep*(r**2))
```

```python
radii = np.arange(0.,20.,0.1)
a = 3
b = 6 # b = 2a
e_mags = [shell_E_field(r,a,b) for r in radii]
```

```python
plt.rcParams['text.usetex'] = True
plt.figure(figsize=(10,5))
plt.plot(radii,e_mags,color='blue')
plt.ylabel(r'$\vec{E}$',fontsize=20)
plt.xlabel("r",fontsize=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.axvline(a,label="a={}".format(a),color='red',linestyle='dashed')
plt.axvline(b,label="b={}".format(b),color='green',linestyle='dashed')
plt.legend(prop={'size': 20})
plt.savefig("/Users/avipatel/Documents/Spring2022/PHYS309/PCS/a2_p3.png")
```
